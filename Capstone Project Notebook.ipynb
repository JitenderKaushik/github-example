```python
import numpy as np
import requests
```

```python
from bs4 import BeautifulSoup

source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text

soup = BeautifulSoup(source, 'html5lib')

#soup
```

```python
ab=soup.find('tr')
tbl=soup.find('table',class_='wikitable sortable')
tbl.tbody.tr.text
```

```python
listKeys=['PostalCode', 'Borough', 'Neighborhood']
df = pd.DataFrame(columns=listKeys)
for table_cell in soup.find_all('tr'):
    listValues=table_cell.text.split("\n")
    if(len(listValues)==5):
        listValues.pop(0)
        listValues.pop(-1)
        di = dict(zip(listKeys, listValues))
        
        if(di["Borough"]!="Not assigned"):
            if(df['PostalCode'].where(df['PostalCode']== di["PostalCode"]).count()>0):  
                df.loc[df['PostalCode'] == di["PostalCode"],["Neighborhood"]] = df.loc[df['PostalCode'] == di["PostalCode"],["Neighborhood"]] +"," + di["Neighborhood"]
            else:
                newRowDF = pd.DataFrame(di, index=[0])
                df = pd.concat([newRowDF, df]).reset_index(drop = True)              
df
```
